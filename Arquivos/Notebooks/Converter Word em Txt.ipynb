# Converte Word em Txt

```
# Importa as bibliotecas necessáriasfrom pathlib import Path  # A forma moderna e orientada a objetos de lidar com caminhos de arquivosfrom docx import Document   # A biblioteca que sabe ler arquivos .docxdef convert_docx_to_txt(docx_file_path: Path) -> str | None:    """    Converte um único arquivo .docx (representado por um objeto Path)    em uma string de texto puro.        Retorna None se ocorrer um erro.    """    print(f"Tentando converter: {docx_file_path.name}")    try:        # Abre o documento Word        document = Document(docx_file_path)                # Cria uma lista vazia para guardar o texto de cada parágrafo        text_lines = []                # Itera sobre cada parágrafo no documento        for paragraph in document.paragraphs:            # Adiciona o texto do parágrafo à nossa lista            text_lines.append(paragraph.text)                    # Junta todas as linhas de texto com uma quebra de linha entre elas        return "\n".join(text_lines)            except Exception as e:        # Se algo der errado (ex: arquivo corrompido), imprime o erro        print(f"!! Erro ao converter o arquivo {docx_file_path.name}: {e}")        return Nonedef main():    """    Função principal que orquestra todo o processo.    """    # --- Parte 1: Definir o caminho e Encontrar os Arquivos ---        # IMPORTANTE: Altere esta variável para a pasta onde estão seus arquivos Word.    # Usei o formato de caminho do Windows como exemplo.    # Path() transforma a string em um objeto de caminho inteligente.    folder_path = Path("/Conversões")        # Verificação: Garante que a pasta realmente existe antes de continuar    if not folder_path.exists():        print(f"!! ERRO: A pasta não foi encontrada: {folder_path}")        print("Por favor, edite o script e ajuste a variável 'folder_path'.")        return  # Para a execução da função    print(f"Iniciando processo na pasta: {folder_path}\n")    # .glob() é um método poderoso para encontrar arquivos que correspondem a um padrão    # Aqui, encontramos todos os arquivos .docx e .doc    print("Procurando arquivos Word (.docx, .doc)...")    files_to_check = list(folder_path.glob('*.docx')) + list(folder_path.glob('*.doc'))        if not files_to_check:        print("Nenhum arquivo Word (.docx ou .doc) encontrado na pasta.")        return    print(f"Arquivos encontrados ({len(files_to_check)}):")    for file_path in files_to_check:        print(f"- {file_path.name}") # .name mostra apenas o nome do arquivo    print("-" * 30) # Imprime uma linha separadora    # --- Parte 2: Processar e Converter os Arquivos ---        # Lista para guardar os dados dos arquivos que conseguimos converter    converted_files_data = []    for file_path in files_to_check:                # Verificamos a extensão do arquivo (em minúsculas)        if file_path.suffix.lower() == '.docx':            # Se for .docx, chamamos nossa função de conversão            text_content = convert_docx_to_txt(file_path)                        # Se a conversão deu certo (não retornou None)            if text_content is not None:                # Guardamos os dados em um dicionário                data = {                    'original_path': file_path,                     'text_content': text_content                }                converted_files_data.append(data)                        elif file_path.suffix.lower() == '.doc':            # Se for .doc, avisamos que não podemos processar            print(f"Aviso: Arquivos .doc (antigos) não são suportados por esta biblioteca. Pulando: {file_path.name}")            print("-" * 30) # Imprime uma linha separadora    # --- Parte 3: Salvar os Novos Arquivos .txt ---    if not converted_files_data:        print("Nenhum arquivo .docx foi convertido com sucesso.")        return    print("Salvando arquivos .txt convertidos...")    for item in converted_files_data:        original_path = item['original_path']        text_content = item['text_content']                # Esta é a mágica do 'pathlib':        # .with_suffix('.txt') troca a extensão original por '.txt'        # Ex: "Relatorio.docx" se torna "Relatorio.txt"        txt_file_path = original_path.with_suffix('.txt')                try:            # .write_text() é a forma simples de salvar um conteúdo de texto            # em um arquivo. 'encoding='utf-8'' garante que acentos            # e caracteres especiais sejam salvos corretamente.            txt_file_path.write_text(text_content, encoding='utf-8')            print(f"Sucesso! Salvo como: {txt_file_path.name}")                    except Exception as e:            # Se der erro ao tentar salvar (ex: permissão negada)            print(f"!! Erro ao salvar o arquivo {txt_file_path.name}: {e}")    print("\nProcesso concluído!")# --- Ponto de Entrada do Script ---# Esta é uma convenção padrão em Python.# O código abaixo só será executado se você rodar este arquivo# diretamente (ex: 'python converter_word.py')if __name__ == "__main__":    main() # Chama nossa função principal
```
