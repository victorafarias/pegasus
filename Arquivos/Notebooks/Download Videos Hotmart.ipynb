```
# BAIXA DO streamyard (vods-storage.streamyard)import requestsfrom tqdm import tqdm # Biblioteca para criar uma barra de progresso (opcional)import osdef baixar_video(url, nome_arquivo_saida="video.mp4", diretorio_saida="/data/Uploads/Downloads"):    """    Baixa um vídeo de uma URL direta e salva localmente com uma barra de progresso.    """    try:        print(f"Iniciando o download de: {url}")        # Faz a requisição para a URL em modo "stream"        # Isso permite baixar o conteúdo em pedaços, sem sobrecarregar a memória        response = requests.get(url, stream=True)        # Verifica se a requisição foi bem-sucedida (código 200)        response.raise_for_status()        # Pega o tamanho total do arquivo para a barra de progresso        tamanho_total = int(response.headers.get('content-length', 0))        # Cria o diretório de saída se não existir        os.makedirs(diretorio_saida, exist_ok=True)        # Define o caminho completo do arquivo de saída        caminho_completo_saida = os.path.join(diretorio_saida, nome_arquivo_saida)        # Configura a barra de progresso com a biblioteca tqdm        barra_progresso = tqdm(            total=tamanho_total,            unit='B',            unit_scale=True,            desc=nome_arquivo_saida        )        # Abre o arquivo local em modo de escrita binária ('wb')        with open(caminho_completo_saida, 'wb') as f:            # Itera sobre o conteúdo do download em pedaços de 8KB            for chunk in response.iter_content(chunk_size=8192):                if chunk: # Filtra chunks vazios                    # Escreve o pedaço no arquivo                    f.write(chunk)                    # Atualiza a barra de progresso                    barra_progresso.update(len(chunk))        # Fecha a barra de progresso ao concluir        barra_progresso.close()        if tamanho_total != 0 and barra_progresso.n != tamanho_total:            print("ERRO: Algo deu errado durante o download.")        else:            print(f"\nDownload concluído! Vídeo salvo como '{caminho_completo_saida}'")    except requests.exceptions.HTTPError as err:        print(f"Erro HTTP: {err}")        print("Verifique se a URL ainda é válida. Links assinados expiram.")    except Exception as e:        print(f"Ocorreu um erro inesperado: {e}")# --- EXECUÇÃO DO SCRIPT ---if __name__ == "__main__":    # Cole a URL do vídeo aqui    url_do_video = "https://vods-storage.streamyard.com/media/g5jnzvhh7id2/vod.mp4?expires=1762791618&signature=791a043387aab24469830b316e53d6d3ba9186458de337648b3ead91bba2a326"    # Defina o nome do arquivo que será salvo    nome_do_arquivo = "Psicologia e Simbólica - Italo Marsili - Aula 5.mp4"    # Define o diretório de saída    diretorio_de_saida = "/content/drive/MyDrive/_Temp"    baixar_video(url_do_video, nome_do_arquivo, diretorio_de_saida)
```
